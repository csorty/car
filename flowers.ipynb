```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def main():
  x, y = load_iris(return_X_y=True)
  #fig, ax = plt.subplots(nrows=4, ncols=4)
  #for i in range(4):
  #  for j in range(4):
  #    ax[i][j].scatter(x[:, i], x[:, j], c=y)
  #plt.show()
  #рисуем проекции на 2 оси из 3д картинки
  #смотрим на 2 характеристики
  #сорта каждый отдельный цвет
  #df = pd.DataFrame(x)
  #df["target"] = y #сорт
  #sns.pairplot(df, hue = "target", palette="dark")
  #plt.show()
  #попарная проекция
  #типо гистограмма, график распределения

  pca = PCA(n_components=2) #трехмероное пространство, только уменьшаем
  pca.fit(x) #x уравнение
  iris2D = pca.transform(x)
  #print(iris2D) #уже не цветок рандомные данные
  #псевдо цветок
  #plt.scatter(iris2D[:, 0], iris2D[:, 1], c=y)
  #plt.show()

  kmeans = KMeans()
  kmeans.fit(x)
  labels = kmeans.labels_
  centroids = kmeans.cluster_centers_
  centroids = pca.transform(centroids)
  plt.scatter(iris2D[:, 0], iris2D[:, 1], c=y)
  plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", color="red") #шумов нет
  plt.show()

if __name__ == "__main__":
    main()
```
