```python
import pandas as pd
import math
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df = pd.read_csv('/content/data.csv')

"""
rows = len(data)
columns = len(data[0])

avg_column = []
for i in range(columns):
    sum_column = 0
    for j in range(rows):
        sum_column += data[j][i]
    avg_column.append(sum_column / rows)

stnd_deviation = []
for i in range(columns):
    sum_sqrt_deviation = 0
    for j in range(rows):
        difference = (data[j][i] - avg_column[i])**2
        sum_sqrt_deviation += difference
    stnd_deviation.append(math.sqrt(sum_sqrt_deviation / rows))

scaled_data = []
for i in range(rows):
    new_row = []
    for j in range(columns):
        if stnd_deviation[j] == 0:
            new_row.append(0)
        else:
            new = ((data[i][j]) - avg_column[j]) / stnd_deviation[j]
            new_row.append(new)
    scaled_data.append(new_row)
"""

means = df.mean()
stnd_deviation = df.std(ddof=0)
scaled = (df - means) / stnd_deviation
scaled_data = scaled.fillna(0)

best_k = 2
max_score = -1
data = scaled_data.values
for k in range(2, 6):
    km = KMeans(n_clusters=k,n_init=10)
    labels = km.fit_predict(data)
    score = silhouette_score(data, labels)
    if score > max_score:
        max_score = score
        best_k = k

kmeans_final = KMeans(n_clusters=best_k, n_init=10)
df['cluster'] = kmeans_final.fit_predict(data)
print(best_k)
df.to_csv('/content/result.csv', index=False)
```
