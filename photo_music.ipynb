```python
import numpy as np
from PIL import Image
import numpy
import librosa
import matplotlib.pyplot as plt

# A прозрачность

def main():
  img = Image.open("img.webp")
  img = np.array(img)
  print(img)

  path = "music.mp3"
  audio = librosa.load(path)
  print(audio)

  path = "траляля"
  y, cr = librosa.load(path, sr = 22050)
  y = y[:10]
  librosa.display.waveshow(y) #точки соединяет
  plt.plot(range(len(y)), y)
  plt.show


if __name__ == '__main__':
  main()

```
